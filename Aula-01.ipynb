## Exemplo 06

```python
url = 'https://guiadacozinha.com.br/category/receitas-de-boteco/'
page = requests.get(url)
recipe = BeautifulSoup(page.text, 'html.parser')
```

```python
my_list = []
for card in recipe.find_all("div",{'class':'recipe-card'}):
    my_list.append(card.find('a')['href'])
```

```python
for i in range(len(my_list)):
    url = my_list[i]
    page = requests.get(url)
    recipe = BeautifulSoup(page.text,'html.parser')
    body = recipe.find("article",{'class':'content-body'})
    print('Título: '+ body.find('h1').text)
    details = recipe.find("div",{'class':'p402_premium'})
    for li in details.find('ul').find_all('li'):
        print(li.text.strip())
    for p in details.find_all('p'):
        print(p.text)
```

## Exemplo 07

```python
url = 'https://en.wikipedia.org/wiki/List_of_FIFA_World_Cup_finals'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
data = []
```

```python
for tr in soup.findAll('table')[2].findAll('tr'):
    cols = tr.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols if ele])
```

```python
print(data)
```

```python
df = pd.DataFrame(data)
```
