## Imports

```
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.chrome.options import Options

import requests
import time
from bs4 import BeautifulSoup
```

## First Example

## Define the options to driver parameters

```
chrome_options = Options()
chrome_options.add_argument("--headless")
```

## Apply the driver to Browser Navigator

```
browser = webdriver.Chrome(options=chrome_options, executable_path="chromedriver_win32/chromedriver.exe")
```

## Attribute the web address to search

```
browser.get('http://www.google.com')
```

## Element input named 'q' receive the key 'dolar' and click on submit button

```
field = browser.find_element_by_name("q")
field.send_keys("dolar")
field.submit()
```

## Printing the content of the site

```
print(browser.page_source)
```

## Saving the current page to a image file

```
browser.get_screenshot_as_file("crawling.jpg")
```

## Closing the browser

```
browser.close()
```

## The same page with BeautifulSoup and Selenium

```
driver = webdriver.Chrome(executable_path="chromedriver_win32/chromedriver.exe")
url = 'https://infomoney.com.br/cryptos/cotacoes'
driver.get(url)
soup = BeautifulSoup(driver.page_source,'html.parser')
table = soup.find('table', attrs={'class':'display border-t-brown-3'})
data = []
table_body = table.find('tbody')
rows = table_body.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols if ele])
    info = [ele for ele in cols if ele]
    post = {
        "index" : info[0],
        "name" : info[1],
        "marketvalue" : info[2],
        "pricers" : info[3],
        "pricebtc" : info[4],
        "total" : info[5],
        "variation" : info[6]
    }
    print(post)
    print("-------------------")

driver.close()
```
